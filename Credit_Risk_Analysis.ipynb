```
# Derived Columns
df['debt_to_income_ratio'] = df['loan_amnt'] / df['person_income']
df['annual_interest_amount'] = (df['loan_amnt'] * df['loan_int_rate']) / 100

def calculate_monthly_installment(principal, annual_rate, years):
    monthly_rate = annual_rate / 12 / 100
    n_payments = years * 12
    installment = (principal * monthly_rate) / (1 - (1 + monthly_rate)**-n_payments)
    return installment

df['loan_term'] = 5  # Assuming a 5-year term for all loans for simplicity
df['monthly_installment'] = df.apply(lambda row: calculate_monthly_installment(row['loan_amnt'], row['loan_int_rate'], row['loan_term']), axis=1)
df['income_left_after_loan'] = df['person_income'] - (df['monthly_installment'] * 12)
df['employment_stability'] = df['person_emp_length'] / df['cb_person_cred_hist_length']
df['credit_utilization'] = df['loan_amnt'] / (df['person_income'] * df['cb_person_cred_hist_length'])
df['loan_to_cred_hist_ratio'] = df['loan_amnt'] / df['cb_person_cred_hist_length']
df['home_ownership_binary'] = df['person_home_ownership'].apply(lambda x: 1 if x == 'OWN' else 0)
loan_grade_mapping = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7}
df['loan_grade_numeric'] = df['loan_grade'].map(loan_grade_mapping)
df['default_history_binary'] = df['cb_person_default_on_file'].apply(lambda x: 1 if x == 'Y' else 0)

# Display the DataFrame with the new columns
print(df)
```
